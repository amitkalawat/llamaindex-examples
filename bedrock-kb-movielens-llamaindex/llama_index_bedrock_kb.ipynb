```python
knowledge_base_id = "8HKONFY8L6"
```

```python
from llama_index.core import Settings
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.tools import QueryEngineTool, ToolMetadata
from llama_index.core.response.pprint_utils import pprint_response

from llama_index.retrievers.bedrock import AmazonKnowledgeBasesRetriever
from llama_index.llms.bedrock import Bedrock
from llama_index.embeddings.bedrock import BedrockEmbedding
```

```python
llm = Bedrock(model = "anthropic.claude-3-sonnet-20240229-v1:0")
embed_model = BedrockEmbedding(model = "amazon.titan-embed-text-v1")
```

```python
Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 1024
```

```python
top_k = 10
search_mode = "HYBRID"
```

```python
bedrock_kb_retriever = AmazonKnowledgeBasesRetriever(
    knowledge_base_id=knowledge_base_id,
    retrieval_config={
        "vectorSearchConfiguration": {
            "numberOfResults": top_k,
            "overrideSearchType": search_mode,
        }
    },
)
bedrock_kb_engine = RetrieverQueryEngine(retriever=bedrock_kb_retriever)
```

```python
query_engine_tools  = [
    QueryEngineTool(
        query_engine=bedrock_kb_engine,
        metadata=ToolMetadata(
            name="bedrock_kb_engine",
            description= ("Provides information about calories across dairy, fruits and vegetables. You can use this tool to find out calories of different ingridients in a recipe."
                          "Use a detailed plain text question as input to the tool."
                         ),
        ),
    ),
    
    QueryEngineTool(
        query_engine=bedrock_kb_engine,
        metadata=ToolMetadata(
            name="bedrock_kb_engine",
            description= ("Provides information about food recipies of various dishes."
                          "Use a detailed plain text question as input to the tool."
                         ),
        ),
    ),
    
        QueryEngineTool(
        query_engine=bedrock_kb_engine,
        metadata=ToolMetadata(
            name="bedrock_kb_engine",
            description= (
                         "Provides information on Amazon, Google, Uber financial data from SEC filings. These are SEC financial documents"
                          "Use a detailed plain text question as input to the tool."
                         ),
        ),
    ),
    
        QueryEngineTool(
        query_engine=bedrock_kb_engine,
        metadata=ToolMetadata(
            name="bedrock_kb_engine",
            description= ("Provides information about movies, actors, plots and any other trivia"
                          "Use a detailed plain text question as input to the tool."
                         ),
        ),
    ),
    
]




```

```python
from llama_index.core.agent import ReActAgent

agent = ReActAgent.from_tools(query_engine_tools, llm=llm, verbose=True)

response = agent.chat("Give me some options to create a grill fish dish which has less than 500 calories?")
print(str(response))
```
